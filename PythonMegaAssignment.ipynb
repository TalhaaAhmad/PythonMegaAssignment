
In this case, the code will print the sum of all the even numbers in the list: 12 + 150 + 180 + 50 = 392.

````python
#CODE

numbers = [12, 75, 150, 180, 145, 525, 50]

# Initialize a variable to store the sum
sum = 0

# Iterate through the numbers in the list
for number in numbers:
    # Check if the number is even
    if number % 2 == 0:
        # Add the number to the sum
        sum += number

# Print the sum
print(sum)
````

```
numbers = [12, 75, 150, 180, 145, 525, 50]
```

To display only those numbers from a list that are divisible by five and meet the other specified conditions, you can use the following code:



This code will iterate through the numbers in the list, and it will use the continue and break statements to skip or stop the loop based on the specified conditions. If the number is divisible by five and meets the other conditions, it will be printed.

````python
numbers = [12, 75, 150, 180, 145, 525, 50]

# Iterate through the numbers in the list
for number in numbers:
    # Check if the number is divisible by five
    if number % 5 == 0:
        # Check if the number is greater than 500
        if number > 500:
            # Stop the loop
            break
        # Check if the number is greater than 150
        if number > 150:
            # Skip the number and move to the next one
            continue
        # Print the number
        print(number)
````

Q76. Write a Python program to find the factorial of a given number.

Here is a Python program that calculates the factorial of a given number using a for loop:

````python
# Function to calculate the factorial of a number
def factorial(n):
    # Initialize the result to 1
    result = 1
    
    # Iterate over the numbers from 1 to n and multiply them
    for i in range(1, n+1):
        result *= i
    
    # Return the result
    return result

# Test the function with different numbers
print(factorial(5))  # Output: 120
print(factorial(10))  # Output: 3628800
print(factorial(15))  # Output: 1307674368000
````

````python
# Function to check if a number is prime
def is_prime(n):
    # Check if the number is less than 2, which is not a prime
    if n < 2:
        return False
    
    # Iterate over the numbers from 2 to the square root of n
    for i in range(2, int(n ** 0.5) + 1):
        # If n is divisible by i, it is not a prime
        if n % i == 0:
            return False
    
    # If the loop completes, the number is a prime
    return True

# Test the function with different numbers
print(is_prime(2))  #
print(is_prime(5))
print(is_prime(9))
````

````python
# Function to check if a number is an Armstrong number
def is_armstrong(n):
    # Convert the number to a string and get the list of digits
    digits = [int(d) for d in str(n)]
    
    # Calculate the sum of the cubes of the digits
    sum_of_cubes = sum(d ** 3 for d in digits)
    
    # Return True if the sum is equal to the number, False otherwise
    return sum_of_cubes == n

# Test the function with different numbers
print(is_armstrong(371))  # Output: True
print(is_armstrong(123))  # Output: False
print(is_armstrong(153))  # Output: True
print(is_armstrong(1634))  # Output: True
print(is_armstrong(456))  # Output: False
````

Q81. Write a Python program to find the n-th Fibonacci Number.

````python
# Function to calculate the n-th Fibonacci number
def fibonacci(n):
    # If n is 0 or 1, return n
    if n == 0 or n == 1:
        return n
    # Otherwise, return the sum of the previous two numbers
    else:
        return fibonacci(n-1) + fibonacci(n-2)

# Test the function with different values of n
print(fibonacci(0))  # Output: 0
print(fibonacci(1))  # Output: 1
print(fibonacci(2))  # Output: 1
print(fibonacci(3))  # Output: 2
print(fibonacci(4))  # Output: 3
print(fibonacci(5))  # Output: 5
print(fibonacci(6))  # Output: 8
print(fibonacci(7))  # Output: 13
print(fibonacci(8))  # Output: 21
print(fibonacci(9))  # Output: 34
````

````python
# Function to find the cumulative sum of a list
def cumulative_sum(lst):
    # Initialize a new list to store the cumulative sum
    cum_sum = []
    # Initialize a variable to store the current sum
    curr_sum = 0
    # Iterate over the elements of the list
    for element in lst:
        # Add the element to the current sum
        curr_sum += element
        # Append the current sum to the cumulative sum list
        cum_sum.append(curr_sum)
    # Return the cumulative sum list
    return cum_sum

# Test the function with different lists
print(f"Cumulative sum: {cumulative_sum([1, 2, 3, 4, 5])}")  # Output: Cumulative sum: [1, 3, 6, 10, 15]
print(f"Cumulative sum: {cumulative_sum([10, 20, 30, 40, 50])}")  # Output: Cumulative sum: [10, 30, 60, 100, 150]
print(f"Cumulative sum: {cumulative_sum([-1, -2, -3, -4, -5])}")  # Output: Cumulative sum: [-1, -3, -6, -10, -15]
print(f"Cumulative sum: {cumulative_sum([0, 0, 0, 0, 0])}")  # Output: Cumulative sum: [0, 0, 0, 0, 0]
````

The function initializes an empty list cum_sum to store the cumulative sum and a variable curr_sum to store the current sum. It then iterates over the elements of the input list and adds each element to curr_sum. The curr_sum is then appended to the cum_sum list and the process is repeated for the remaining elements. Finally, the cum_sum list is returned as the result.

Q86. Write a Python program to check if a string is palindrome or not.

Here is a Python program that checks if a string is a palindrome or not:

````python
# Function to check if a string is a palindrome
def is_palindrome(string):
    # Reverse the string
    reversed_string = string[::-1]
    # Compare the reversed string with the original string
    if string == reversed_string:
        return True
    else:
        return False

# Test the function with different strings
print(f"Is palindrome: {is_palindrome('racecar')}")  # Output: Is palindrome: True
print(f"Is palindrome: {is_palindrome('hello')}")  # Output: Is palindrome: False
print(f"Is palindrome: {is_palindrome('madam')}")  # Output: Is palindrome: True
print(f"Is palindrome: {is_palindrome('level')}")  # Output: Is palindrome: True
print(f"Is palindrome: {is_palindrome('redivider')}")  # Output: Is palindrome: True
````

The function first reverses the string using slicing and then compares the reversed string with the original string. If they are equal, it returns True, otherwise it returns False.

Q87. Write a Python program to remove i'th element from a string.

````python
# Function to remove the i'th element from a string
def remove_ith_element(string, i):
    # Convert the string to a list of characters
    char_list = list(string)
    # Delete the i'th element from the list
    del char_list[i]
    # Join the list of characters to form a new string
    new_string = "".join(char_list)
    # Return the new string
    return new_string

# Test the function with different strings and indices
print(f"New string: {remove_ith_element('abcdef', 0)}")  # Output: New string: bcdef
print(f"New string: {remove_ith_element('abcdef', 2)}")  # Output: New string: abdef
print(f"New string: {remove_ith_element('abcdef', 4)}")  # Output: New string: abcde
print(f"New string: {remove_ith_element('abcdef', 5)}")  # Output: New string: abcde
````

The function first converts the string to a list of characters using the list() function. It then uses the del statement to delete the i'th element from the list. Finally, it joins the list of characters using the join() function to form a new string and returns it.

Q89. Write a Python program to find words which are greater than given length k.

Here is a Python program that finds words in a string which are greater than a given length k:

````python
# Function to find words which are greater than given length k
def find_long_words(string, k):
    # Split the string into a list of words
    words = string.split()
    # Initialize an empty list to store the long words
    long_words = []
    # Iterate over the list of words
    for word in words:
        # If the length of the word is greater than k, add it to the list of long words
        if len(word) > k:
            long_words.append(word)
    # Return the list of long words
    return long_words

# Test the function with different strings and values of k
print(f"Long words: {find_long_words('abc def ghi', 2)}")  # Output: Long words: ['abc', 'def', 'ghi']
print(f"Long words: {find_long_words('abc def ghi', 3)}")  # Output: Long words: ['abc', 'def']
print(f"Long words: {find_long_words('abc def ghi', 4)}")  # Output: Long words: ['abc']
print(f"Long words: {find_long_words('abc def ghi', 5)}")  # Output: Long words: []
````

The function first splits the string into a list of words using the split() function. It then initializes an empty list to store the long words and iterates over the list of words. For each word, it checks if the length of the word is greater than k using the len() function. If it is, it adds the word to the list of long words. Finally, it returns the list of long words.

Here is a Python program that extracts unique dictionary values:

````python
# Function to extract unique dictionary values
def extract_unique_values(dictionary):
    # Initialize an empty set to store the unique values
    unique_values = set()
    # Iterate over the dictionary values
    for value in dictionary.values():
        # Add the value to the set of unique values
        unique_values.add(value)
    # Return the set of unique values
    return unique_values

# Test the function with different dictionaries
print(f"Unique values: {extract_unique_values({'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5})}")  # Output: Unique values: {1, 2, 3, 4, 5}
print(f"Unique values: {extract_unique_values({'a': 1, 'b': 2, 'c': 3, 'd': 3, 'e': 5})}")  # Output: Unique values: {1, 2, 3, 5}
print(f"Unique values: {extract_unique_values({'a': 1, 'b': 2, 'c': 3, 'd': 3, 'e': 3})}")  # Output: Unique values: {1, 2, 3}
````

Here is a Python program that demonstrates how to merge two dictionaries:

````python
# Function to merge two dictionaries
def merge_dictionaries(dict1, dict2):
    # Initialize a new dictionary to store the merged dictionary
    merged_dict = {}
    # Iterate over the keys and values of the first dictionary
    for key, value in dict1.items():
        # Add the key-value pair to the merged dictionary
        merged_dict[key] = value
    # Iterate over the keys and values of the second dictionary
    for key, value in dict2.items():
        # Add the key-value pair to the merged dictionary
        merged_dict[key] = value
    # Return the merged dictionary
    return merged_dict

# Test the function with different dictionaries
print(f"Merged dictionary: {merge_dictionaries({'a': 1, 'b': 2, 'c': 3}, {'d': 4, 'e': 5})}")  # Output: Merged dictionary: {'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5}
print(f"Merged dictionary: {merge_dictionaries({'a': 1, 'b': 2, 'c': 3}, {'c': 4, 'd': 5})}")  # Output: Merged dictionary: {'a': 1, 'b': 2, 'c': 4, 'd': 5}
print(f"Merged dictionary: {merge_dictionaries({'a': 1, 'b': 2, 'c': 3}, {})}")  # Output: Merged dictionary: {'a': 1, 'b': 2, 'c': 3}
````

The function initializes a new dictionary to store the merged dictionary and iterates over the keys and values of the first dictionary. For each key-value pair, it adds the pair to the merged dictionary using the assignment operator. Then, it iterates over the keys and values of the second dictionary and adds the pairs to the merged dictionary in the same way. Finally, it returns the merged dictionary.

Q94. Write a Python program to get all combinations of 2 tuples.

Input : test_tuple1 = (7, 2), test_tuple2 = (7, 8)

Output : [(7, 7), (7, 8), (2, 7), (2, 8), (7, 7), (7, 2), (8, 7), (8, 2)]

````python
# Function to get all combinations of 2 tuples
def get_combinations(tuple1, tuple2):
    # Initialize a new list to store the combinations
    combinations = []
    # Iterate over the elements in the first tuple
    for elem1 in tuple1:
        # Iterate over the elements in the second tuple
        for elem2 in tuple2:
            # Append a tuple with the current elements to the list
            combinations.append((elem1, elem2))
    # Return the list of combinations
    return combinations

# Test the function with the given tuples
test_tuple1 = (7, 2)
test_tuple2 = (7, 8)
print(f"Combinations: {get_combinations(test_tuple1, test_tuple2)}")  # Output: Combinations: [(7, 7), (7, 8), (2, 7), (2, 8)]
````
